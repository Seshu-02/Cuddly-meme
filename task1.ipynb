```python
import matplotlib.pyplot as plt

# Data from the image
age_groups = ['0-20 Years', '21-64 Years', '65+ Years']
population = [512, 807, 98]
colors = ['#FFE600', '#00A1FF', '#FF2F92']  # matching image colors

plt.figure(figsize=(8,3 ))
plt.bar(age_groups, population, color=colors)
plt.title("India's Population Distribution by Age (2022)")
plt.xlabel("Age Group")
plt.ylabel("Population (Millions)")
plt.ylim(0, 900)
for i, v in enumerate(population):
    plt.text(i, v + 20, f"{v}M", ha='center', fontweight='bold')
plt.show()

```
